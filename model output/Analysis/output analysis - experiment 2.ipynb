```python
import matplotlib.pyplot as plt
import numpy as np

# Dati
categories = ['Old Model Replies', 'Neither', 'Young Model Replies']
young_responses = [154, 141, 122]
old_responses = [132, 175, 110]

# Larghezza delle barre
bar_width = 0.35

# Posizioni delle barre
r1 = np.arange(len(categories))
r2 = [x + bar_width for x in r1]

# Creazione delle barre con colori più accattivanti
plt.bar(r1, young_responses, color='#1f77b4', width=bar_width, label='Young')
plt.bar(r2, old_responses, color='#ff7f0e', width=bar_width, label='Old')

# Aggiunta di etichette e titolo
plt.xlabel('Replies', fontweight='bold')
plt.ylabel('Number of Responses', fontweight='bold')
plt.title('Responses by Age Group')
plt.xticks([r + bar_width/2 for r in range(len(categories))], categories)

# Aggiunta di legenda
plt.legend()

# Mostra il grafico
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dati forniti
categories = ['O-Model', 'Y-Model']
young_responses = [154, 122]
old_responses = [132, 110]

# Costruzione della matrice di confusione con le righe invertite
confusion_matrix = np.array([
    young_responses,  # Righe per Young individuals
    old_responses     # Righe per Old individuals
])

# Creazione del grafico
plt.figure(figsize=(8, 6))
plt.imshow(confusion_matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.colorbar()

# Aggiunta delle etichette agli assi
tick_marks = np.arange(len(categories))
plt.xticks(tick_marks, categories, rotation=45, fontsize=16)  # Aumenta la dimensione del font a 20
plt.yticks(tick_marks, ['Young Individuals', 'Old Individuals'], fontsize=16)  # Aumenta la dimensione del font a 20

# Aggiunta dei valori all'interno dei rettangoli
for i, j in np.ndindex(confusion_matrix.shape):
    plt.text(j, i, format(confusion_matrix[i, j], 'd'),
             ha="center", va="center",
             fontsize=16,
             color="white" if confusion_matrix[i, j] > 122 else "black")  # Forza un contrasto maggiore

plt.tight_layout()

plt.show()
```

```python
import matplotlib.pyplot as plt

# Dati
categories = [
    'Complete match',
    'Complete mismatch',
    'Partial match',
    'Partial mismatch',
    'Agreement'
]

values = [32, 31, 77, 89, 113]
percentages = [7.67, 7.43, 18.47, 21.34, 27.10]

# Colori personalizzati per le barre
colors = ['#1f77b4', '#1f77b4', '#1f77b4', '#1f77b4', '#1f77b4']

# Creazione del grafico a barre
plt.figure(figsize=(10, 6))
bars = plt.bar(categories, values, color=colors)

# Aggiunta delle etichette delle percentuali sopra le barre
for bar, percentage in zip(bars, percentages):
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 2, f'{percentage}%', ha='center', va='bottom')

# Aggiunta di etichette e titolo
plt.xlabel('Categories', fontweight='bold', fontsize=14)
plt.ylabel('Count', fontweight='bold', fontsize=14)
plt.title('Distribution of Responses', fontsize=16)
plt.xticks(rotation=0, fontsize=12)  # Aumentare la dimensione del font delle etichette delle categorie

# Mostra il grafico
plt.tight_layout()
plt.show()
```
