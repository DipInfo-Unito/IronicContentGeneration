```python
import pandas as pd
import krippendorff
import numpy as np
```

```python
df = pd.read_csv("/home/marem/VscProjects/theGIRLS/annotation/E1_annotations_clean.csv", sep=",")
print(df.shape)
```

```python
df
```

```python
df_iaa = df.drop(["StartDate", "EndDate", "Status", "IPAddress", "Progress", "Duration (in seconds)",
                  "Finished", "RecordedDate", "ResponseId", "RecipientLastName", "all_attentions",
                  'RecipientFirstName', 'RecipientEmail', 'ExternalReference',
                  'LocationLatitude', 'LocationLongitude', 'DistributionChannel',
                  'UserLanguage', 'Q801', 'Q1108'], axis=1)

df_iaa=df_iaa.set_index("PROLIFIC_PID")
df_iaa
```

```python
df_label_matrix = df_iaa.replace({"Strongly Disagree": -1, "Disagree": -1, "Neither Agree nor Disagree": 0, "Agree": 1, "Strongly Agree": 1})
label_matrix = df_label_matrix.to_numpy(na_value=np.nan, dtype='float')

iaa_krip = krippendorff.alpha(reliability_data=label_matrix, level_of_measurement='nominal')

print (f"{iaa_krip:.2f}")
```
